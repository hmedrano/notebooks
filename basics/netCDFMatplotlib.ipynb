```python
# Graficos inline en el notebook
%matplotlib inline  

plt.figure(figsize=(10,6), dpi=80)   # Tamano de la figura en pulgadas, dpi es el numero de puntos por pulgada

axt = plt.subplot(211)
axt.plot(temp)
axt.set_xlim(0,temp.size)

axs = plt.subplot(212)
axs.plot(sal)
axs.set_xlim(0,sal.size)
```

```python
# Graficar con mas informacion

# Convertir dimension temporal a tipo de dato datetime
ptimes = nc.num2date(time[:],time.units)

plt.figure(figsize=(10,6), dpi=80)
plt.plot_date(ptimes[0:-1:10],temp[0:-1:10], "k-", label=temp.long_name)   # Metodo para incluir fechas en alguno de los ejes

plt.legend(loc='upper right') # Mostrar leyenda y definir su ubicacion

plt.ylabel(temp.units)        # Etiqueta para el eje y

plt.title('Datos instrumento MCT')   
```

```python
# Anotaciones
fig = plt.figure(figsize=(10,6), dpi=80)
ax = fig.gca()   # Obtener el "Axes" axtual, o crear uno!

ax.plot_date(ptimes[:],temp[:], "r-", label=temp.long_name )
ax.legend(loc='upper right')
ax.set_ylabel(temp.units)
ax.set_title('Datos instrumento MCT')

ax.annotate(r' $\min(temp) $' , xy=( ptimes[temp[:].argmin()], temp[:].min() ) , 
            xycoords='data', fontsize = 14, color='k')

```

### Graficado de datos en mallas, y sobre mapas
Usando el la libreria Basemap

```python
# Importando la funcionalidad
from mpl_toolkits.basemap import Basemap
import numpy as np

fig = plt.figure(figsize=(10,6), dpi=80)
# Generamos un mapa, con alguna proyeccion.
map = Basemap(projection='robin',lat_0=25,lon_0=-100,resolution='l')

# Dibujar linea costena y rellenar continentes.
map.drawcoastlines(linewidth=0.25)
map.fillcontinents(color='gray',lake_color='blue')

# rellenar de color el oceano
map.drawmapboundary(fill_color='royalblue')

# Dibujar lineas de meridianos y paralelos cada 30 grados
map.drawmeridians(np.arange(0,360,30))
map.drawparallels(np.arange(-90,90,30))

plt.title('contour lines over filled continent background')
```

```python
from mpl_toolkits.basemap import interp

# Cargar datos del modelo numerico
dst = nc.Dataset('datasets/GLORYS2V3_1dAV_20121010_20121011_gridT_R20121017.nc','r')
tempsurface = dst.variables['votemper'][0,0,:,:]
lon_curv = dst.variables['nav_lon'][:]
lat_curv = dst.variables['nav_lat'][:]

fig = plt.figure(figsize=(10,6), dpi=80)
# Generamos un mapa, con alguna proyeccion.
#map = Basemap(projection='robin',lat_0=15,lon_0=-120,resolution='l')
#map = Basemap(projection='hammer',lat_0=15,lon_0=-120,resolution='l')
map = Basemap(projection='ortho',lat_0=15,lon_0=-120,resolution='l')

# Dibujar linea costena y rellenar continentes.
map.drawcoastlines(linewidth=0.25)
map.fillcontinents(color='gray',lake_color='blue')
# Rellenar los oceanos
map.drawmapboundary(fill_color='royalblue')

# Generar posiciones x y y, en base a la proyeccion del mapa.
x, y = map(lon_curv, lat_curv)

cs = map.contourf(x,y, tempsurface, levels = np.arange(tempsurface.min(),tempsurface.max(),0.2) , extend='both')
```

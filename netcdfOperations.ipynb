```python
import netCDF4 as nc 
import numpy as np

datasetName = 'crudosFNL_2014-11-26__2014-11-30.nc'
dst = nc.Dataset(datasetName,'r') 

print ':: Listado de dimensiones :: '
for d in dst.dimensions.itervalues():
    print d 

print ':: Listado de variables :: '
for v in dst.variables.itervalues():
    print v 
    print '\n'
```

```python
# Obtener valores maximos para U y V en cada paso de tiempo

MaxValues = []

for ts in range(0,len(dst.dimensions['time'])):
    [iu, ju] = np.unravel_index (np.argmax(dst.variables['ugrd10m'][ts,:,:]) , dst.variables['ugrd10m'][ts,:,:].shape)
    print 'Indices valor Maximo U : ' + str([iu,ju])
    print 'Valor Maximo U : ' + str(np.max(dst.variables['ugrd10m'][ts,iu,ju]))
    
    [iv, jv] = np.unravel_index (np.argmax(dst.variables['vgrd10m'][ts,:,:]) , dst.variables['vgrd10m'][ts,:,:].shape)
    print 'Indices valor Maximo V : ' + str([iv,jv])
    print 'Valor Maximo V : ' + str(np.max(dst.variables['vgrd10m'][ts,iv,jv]))    
    MaxValues.append([ np.max(dst.variables['ugrd10m'][ts,iu,ju]) , np.max(dst.variables['vgrd10m'][ts,iv,jv]) ])
    
# Cerrar el archivo netcdf    
dst.close()
```

```python
# Graficar los valores maximos: MaxValues
%matplotlib inline
import matplotlib.pyplot as plt

plt.plot(np.array(MaxValues))
```

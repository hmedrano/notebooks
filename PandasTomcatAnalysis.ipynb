## Script en python para generar graficos de visitas y peticiones del servidor tomcat.  Usando pandas, matplotlib y apachelog para analizar los archivos de log generados.

```python
import os
import apachelog
import numpy as np 
import matplotlib.pyplot as plt 
import pandas as pd 
from glob import glob

%pylab inline

logPath = 'sample_logs'
nFormatOld = r'%h %l %u %t \"%r\" %>s %b '
nFormat = r'%h %l %u %t \"%r\" %>s %b \"%{User-Agent}i\"'
myParserOld = apachelog.parser(nFormatOld) 
myParser = apachelog.parser(nFormat) 

# Get localhost_access files
logs = glob(os.path.join(logPath,'localhost_access*.txt'))
logs.sort() 
# log files to analize
print logs
```

```python
log_list = []
# Parse all files
for logfile in logs:    
    fp = open(logfile,'r')
    
    for line in fp.readlines():
        data = None 
        try:
            data = myParserOld.parse(line)  
        except Exception, e:
            try:
                data = myParser.parse(line)
            except Exception ,e:
                # print 'No se pudo hacer el parse de : ' + str(line) 
                continue
        data['%t'] = data['%t'][1:12]+' '+data['%t'][13:21]+' '+data['%t'][22:27]
        # Not include the cic-pem, and my ip.
        if (data['%h'] != '158.97.1.9' and data['%h'] != '158.97.83.160' and data['%h'] != '0:0:0:0:0:0:0:1'):
            log_list.append(data)

    fp.close()
df = pd.DataFrame(log_list) 
df = df.rename(columns={'%h':'host' , '%l':'identd' , '%u':'User' , '%t':'Time' , '%r':'Request' , '%>s':'Status' , '%b':'Size' , '%{User-Agent}i':'User-Agent'})

# Make index 'Time' columnn
df.index = pd.to_datetime(df.pop('Time'))

def dash2nan(x):
    if x == '-':
        x = 0
    else:
        x = float(x)/1048576.
    
    return x

df['Size'] = df['Size'].apply(dash2nan)
print df['Size']
```

```python
# Plot Trafic - outgoing
df['Size'].plot()
```

```python
# Count unique hosts 
print df['host'].value_counts()
```

```python
# Peticiones diarias y MB descargados diariamente.
df_b = df['Size'].resample('D', how=['count',np.sum])
df_b['count'].plot( color='r')
legend('Cuenta')
df_b['sum'].plot(secondary_y=True)
legend('Total MB')
```

```python
# Count unique daily visits
dailyU = df.groupby(df.index.date)['host'].nunique()
dailyU.plot(kind='bar',label='Visitas unicas')
legend()
```

```python
pd.options.display.mpl_style = 'default'
# Plot unique daily visits and total hits

dailyU = df.groupby(df.index.date)['host'].nunique()
dailyU.index = dailyU.index.to_datetime()
dailyU = dailyU.resample('D')
dailyU.plot(label='Visitas unicas',legend=True)
plt.legend(loc='upper left')
df_b = df['host'].resample('D', how='count')
df_b.plot(color='r', secondary_y=True, label='Peticiones',legend=True)
plt.legend(loc='upper right')
```
